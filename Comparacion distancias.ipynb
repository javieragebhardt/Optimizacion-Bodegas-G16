```python
import construccion_datos
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
d_Manhattan = construccion_datos.d_Manhattan
d_mapbox = construccion_datos.d_mapbox
```

```python
def calcular_estadisticas_y_graficos(diccionario, nombre_diccionario):
    lista_distancias = [dist for distances in diccionario.values() for dist in distances.values()]
    
    promedio_distancia = np.mean(lista_distancias)
    distancia_minima = np.min(lista_distancias)
    distancia_maxima = np.max(lista_distancias)
    desviacion_distancia = np.std(lista_distancias)
    
    # Estadísticas 
    print(f'Estadísticas para {nombre_diccionario}:')
    print(f'Promedio de distancias: {promedio_distancia}')
    print(f'Distancia mínima: {distancia_minima}')
    print(f'Distancia máxima: {distancia_maxima}')
    print(f'Desviación estándar: {desviacion_distancia}\n')

    # Histograma
    plt.hist(lista_distancias, bins=20, edgecolor='k', alpha=0.5, label=nombre_diccionario)
```

```python
# Valores para ambos diccionarios
calcular_estadisticas_y_graficos(d_mapbox, 'd_mapbox')
calcular_estadisticas_y_graficos(d_Manhattan, 'd_Manhattan')

# Gráfico
plt.xlabel('Distancia')
plt.ylabel('Frecuencia')
plt.title('Comparación de Distribución de Distancias')
plt.legend()
plt.show()
```
